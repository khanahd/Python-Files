```python
import datetime
import pytz
```

```python
class Account:
    """ Simple account class with balance """

    @staticmethod   # static method
    def _current_time(self):
        utc_time = datetime.datetime.utcnow()
        return pytz.utc.localize(utc_time)

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transaction_list = []
        print("Account created for {}".format(self.name))


    def deposit(self, amount):
        if amount > 0:
            self.balance +=amount
            self.transaction_list.append((pytz.utc.localize(datetime.datetime.utcnow()),amount))
        print("Current Balance: ", self.balance)

    def withdraw(self,amount):
        if amount==0:
            print('No amount to withdraw!')
        elif amount<=self.balance:  # to prevent the balance from going negative
            self.balance -= amount
            self.transaction_list.append((pytz.utc.localize(datetime.datetime.utcnow()),amount))
        else:
            print("Can't withdraw {} as the balance does not allow it.".format(amount))
        print("Current Balance: ", self.balance)


    def show_balance(self):
        print("Balance is : ",self.balance)

    def show_transactions(self):
        for date,amount in self.transaction_list:
          if amount > 0:
            tran_type = 'deposited'
          else:
            tran_type = 'withdrawn'
            amount *= -1
          print("{:6} {} on {} (local time was {})".format(amount,tran_type,date,date.astimezone()))

```

```python
hdfc = Account('ahad',5000)

hdfc.deposit(100)
```

```python
hdfc.show_balance()

hdfc.deposit(83424)

hdfc.show_balance()

hdfc.withdraw(43567)
hdfc.withdraw(00)
hdfc.withdraw(50000)
```
